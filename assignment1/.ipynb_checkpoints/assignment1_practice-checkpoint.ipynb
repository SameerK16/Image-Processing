```python
time1 = time.time()
img = image.imread('Tiger.jpg')
row, col, ch  = img.shape
img = np.reshape(img, [row*col, ch])

img1 = np.zeros([row*col, ch])
img1[:,0] = 0.3*img[:,0]
img1[:,1] = 0.59*img[:,1]
img1[:,2] = 0.11*img[:,2]

img_g4 = [i[0] + i[1]+ i[2] for i in img1 ]
img_g4 = np.reshape(img_g4, [row, col])
print(time.time() - time1)
```

```python
def gray2bin(img1, threshold):
    
    row, col= img1.shape
    img_copy = np.reshape(img1, [row*col])
    bin1 = lambda x, threshold: 1 if x > threshold else 0  
    
    img_bin = np.zeros([row, col])
    img_bin = np.asarray([bin1(i, threshold) for i in img_copy])
    img_bin = np.reshape(img_bin, [row, col])
    
    plt.imshow(img_bin, cmap = plt.get_cmap('gray'))
    plt.show()
```

```python
gray2bin(img_g4, 127)
```
